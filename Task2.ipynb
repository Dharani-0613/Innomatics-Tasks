```python
#Unique words in a sentence
sentence = input("Enter a sentence: ")

#Split into words
words= sentence.split()
unique_words= set(words)

#Convert list to set to remove duplicates
unique_words= set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
employees = { "Ravi" : 75000, "Anita" : 68000, "Kiran" : 72000 }
highest_employee = None
highest_salary = 0

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name
print ("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#Find maximum and minimum values
numbers = [45,22,89,10,66]
#Initialize max and min with first element
max= numbers[0]
min= numbers[0]

for num in numbers:
    if num>max:
        max = num
    if num < min:
        min = num
print ("List:", numbers)
print ("Max:", max)
print ("Min:", min)
```

```python
#Count products above a price threshold
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print ("Products above 1000:", count)
```

```python
#Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
present_count = 0
for status in attendance:
    if status == "P":
        present_count += 1
total_days = len(attendance)
percentage = (present_count / total_days) * 100

print("Attendance Percentage: ", percentage)
```

```python
#Count character frequency
text = "python"
char_count = {}
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
print(char_count)
```

```
#Smart parking lot management system
def parking(capacity, logs):
  parked = 0
  peak = 0
  for log in logs:
    if log == "IN" :
      parked += 1
    elif log == "OUT" :
      parked -= 1
    #Peak usage
    peak = max(peak, parked)
  status = "Full" if parked >= capacity else "Available"
  print(f"Currently Parked Vehicles:",parked)
  print(f"Parking Status:",status)

#Example
parking(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```
#online food delivery time estimator
def del_time(dist, traffic, weather):
  base_time = dist * 5 #taking 5 min per km
  if traffic.lower() == "high":
    base_time += 10
  elif traffic.lower() == "medium":
    base_time += 5
  if weather.lower() in ["rainy","stormy"]:
    base_time += 5
  print(f"Estimated Delivery Time: {base_time} minutes")

#Example
del_time(8, "High", "Rainy")
```

```
#movie theatre set occupancy analyzer
def seat_occ(tot_seats, booked):
  booked_count = len(booked)
  #percentage of occupancy
  occupancy = (booked_count / tot_seats) * 100
  if occupancy >= 100:
    status = "Housefull"
  elif occupancy >= 75:
    status = "Almost Full"
  else:
    status = "Available"
  print(f"Occupancy: {occupancy:.0f}%")
  print(f"Show Status: {status}")

#Example
seat_occ(200, [1]*150)
```

```
#Cloud server load classification system
def server_load(cpu_read):
  #calculate average
  avg_load = sum(cpu_read)/len(cpu_read)
  #conditions
  if avg_load < 50:
    status = "Normal"
  elif 50 <= avg_load < 80:
    status = "Warning"
  else:
    status = "Critical"
  print(f"Average CPU Load: {avg_load:.0f}%")
  print(f"Server Status: {status}")

#Example
server_load([45,60,70,85,90])
```

```
#Smart Classroom Resource usage Monitor
def res_monitor(usage):
  overused = [r for r, h in usage.items() if h > 8]
  alert = "Yes" if overused else "No"
  print(f"Overused Resources: {', '.join(overused) if overused else 'None'}")
  print(f"Energy Alert: {alert}")

#Example
res_monitor({"Projector": 6, "AC": 9, "Lights": 4})
```

```
#Online event registration capacity controller
def reg_controller(capacity, logs):
  confirmed = min(capacity, logs)
  waitlisted = max(0, logs - capacity)
  status = "Closed" if logs > capacity else "Open"
  print(f"Confirmed Registrations: {confirmed}")
  print(f"Waitlisted Registrations: {waitlisted}")
  print(f"Registration Status: {status}")

#Example
reg_controller(100,105)
```

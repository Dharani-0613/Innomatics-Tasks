```python
#Smart Light Controller
class smart_light:
  def __init__(self,light_name,action):
    self.light_name = light_name
    self.action = action

  def display(self):
    print(f"{self.light_name} is {self.action}")

light_name= input("Light Name:")
action= input("Action:")

light1 = smart_light(light_name, action)
light1.display()
```

```python
#Employee ID Card System
class employee:
  def __init__(self,name,id,dept):
    self.name= name
    self.id= id
    self.dept= dept

  def display(self):
    print("Employee ID Card")
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Department: {self.dept}")

name= input("Employee name:")
id= input("Employee ID: ")
dept= input("Department: ")

emp = employee(name, id, dept)
emp.display()
```

```python
#Mobile contact record
class contact_info:
  def __init__(self, name, phn_no):
    self.name= name
    self.phn_no= phn_no

  def display(self):
    print("Contact saved")
    print(f"Name: {self.name}")
    print(f"Phone: {self.phn_no}")

name= input("Contact Name:")
phn_no= input("Phone Number:")

exmple= contact_info(name,phn_no)
exmple.display()
```

```python
#Product price tag generator

class product:
  def __init__(self,prod_name, prod_price):
    self.prod_name= prod_name
    self.prod_price= prod_price

  def display(self):
    print(f"Product: {self.prod_name}")
    print(f"Price: ₹{self.prod_price}")

prod_name= input("Product Name:")
prod_price= input("Price:")

prod1 = product(prod_name, prod_price)
prod1.display()
```

```python
#Movie rating display system
class movie:
  def __init__(self,movie_name, rating):
    self.movie_name= movie_name
    self.rating= rating

  def display(self):
    print(f"Movie Name: {self.movie_name}")
    print(f"Rating: {self.rating}/5")

movie_name= input("Movie Name:")
rating= input("Rating:")

movie1 = movie(movie_name, rating)
movie1.display()
```

```python
#Delivery Address Manager
class delivery_details:
  def __init__(self,customer_name,address):
    self.customer_name= customer_name
    self.address= address

  def display(self):
    print("Delivery Details")
    print(f"Customer: {self.customer_name}")
    print(f"Address: {self.address}")

customer_name= input("Customer Name:")
address= input("Address:")
cust1= delivery_details(customer_name,address)
cust1.display()
```

```python
#Employee performance bonus eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
high_score = max(employees.values())
#finding all employees with high score
top_performers = []
for name, score in employees.items():
    if score == high_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ",".join(top_performers),f"(Score: {high_score})")
```

```python
#search query keyword analysis
import string
query = "BUy mobile phone buy phone online"
#Converting to lowercase
query = query.lower()
#removing punctutation
query = query.translate(str.maketrans('','',string.punctuation))

words = query.split()
frequency = {}
for word in words:
  frequency[word] = frequency.get(word,0) + 1
#filter words appearing more than oncce
repeated_words = {word: count for word, count in frequency.items() if count > 1}

print(repeated_words)
```

```python
#Sensor Data Validation
sensor_readings = [3,4,7,8,10,12,5]
valid_readings = []
for index, value in enumerate(sensor_readings):
  if value % 2 == 0:
    valid_readings.append((index,value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}
#Extracting domains and count
for email in emails:
  domain = email.split('@')[1]
  domain_count[domain] = domain_count.get(domain, 0) + 1

#calculate percentage
total_emails = len(emails)
for domain, count in domain_count.items():
  percentage = (count / total_emails) * 100
  print(f"{domain}: {percentage: .0f}%")
```

```python
#Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]
#calculating average and threshold = 30% above avg
avg_sales = sum(sales)/ len(sales)
threshold = avg_sales * 1.3
#detect spike days
for index, value in enumerate(sales):
  if value > threshold:
    print(f"Day {index + 1}: {value}")
```

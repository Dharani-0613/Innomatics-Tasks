```python
#Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
#counting the total pass students & fail students
for mark in marks:
    if mark >=50 :
        pass_count +=1
    else:
        fail_count +=1

print("Total number of pass students:",pass_count)
print("Total number of fail students:",fail_count)
```

```python
# Simple Data Cleaner
names = [" Alice ","bob", " CHARLIE "]
#removing extraspaces using "strip" & converting into lowercase
clean_list = [name.strip().lower() for name in names]
#printing the cleaned list
print (clean_list)
```

```python
# Message Length Analyzer
messages = ["Hi","Welcome to the platform","OK"]

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    #if length is longer than 10 characters- flag it
    if length > 10:
        print("Status: FLAGGED")
    else:
        print("Status: OK")
```

```python
#Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
    if log == "ERROR":
        error_count +=1

print("Total error count:", error_count)
```
